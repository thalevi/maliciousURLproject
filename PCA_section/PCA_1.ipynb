```python
##using PCA to visualization
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('All.csv')
dataset.shape
```

```python
##delete Nan values and seperate x and y
# y as result
dataset = pd.DataFrame(dataset)
dataset.dropna(inplace = True)
x = dataset.iloc[:, 0:79].values
y = dataset.iloc[:, 79].values
```

```python
##StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x)
```

```python
##import PCA set components =40

from sklearn.decomposition import PCA
pca = PCA(n_components=40)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents)
```

```python
principalDf
```

```python
y=pd.DataFrame(y)
```

```python
finalDf = principalDf.iloc[:,0:2]
finalDf
```

```python
## add new column 'URL_Type_obf_Type'
finalDf = pd.concat([finalDf, y], axis = 1)

## delete Nan values
finalDf.dropna(inplace = True)
finalDf.columns=['principal component 1', 'principal component 2','URL_Type_obf_Type']
finalDf
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Defacement', 'malware', 'phishing','spam','benign']
colors = ['r', 'g', 'b','y','k']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['URL_Type_obf_Type'] == target
    ax.scatter(finalDf.loc[indicesToKeep,'principal component 1']
               , finalDf.loc[indicesToKeep,'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
