```python
import pandas as pd
import numpy as np 
from sklearn.ensemble import RandomForestRegressor
```

```python
dataset  = pd.read_csv('All.csv')
```

```python
dataset.shape
```

0: benign  1:defacement   2: phishing 3: malware  4:spam


```python
dataset.loc[dataset["URL_Type_obf_Type"] !='benign', "URL_Type_obf_Type"] = 1
dataset.loc[dataset["URL_Type_obf_Type"] =='benign', "URL_Type_obf_Type"] = 0
```

```python
dataset = pd.DataFrame(dataset)
##drop rows containing NaN value

dataset.dropna(inplace = True)
print(dataset.shape)
dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 79].values
y=y.astype('int')
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2,
                                                random_state=0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#n_estimators 10-100: default 100
regressor = RandomForestRegressor(70)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
y_pred=np.round(y_pred,0)
```

```python

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
